```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sepal_length = df['sepal.length']
```

```python
df['variety'].unique()

```

```python
plt.figure(figsize=(8,6))
plt.hist(sepal_length,bins = 20,alpha=0.6,density='True',color='r',edgecolor='black')
plt.title('Distribution of Sepal Legth')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(sepal_length,color='r',fill=True,label="PDF")
plt.title('Distribution of Sepal Legth')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
sorted_sepal_length = np.sort(sepal_length)
cdf = np.arange(1,len(sorted_sepal_length)+1)/len(sorted_sepal_length)

plt.figure(figsize=(8,6))
plt.plot(sorted_sepal_length,cdf,marker='.',linestyle='none',color='b')
plt.title('CDF of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```
